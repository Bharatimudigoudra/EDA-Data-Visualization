```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/diabetes.csv')
print(data.head())
```

```python
# Summary statistics of numerical features
data.describe()
```

```python
# Data types of all columns
data.dtypes
```

```python
# Checking for missing values
data.isnull().sum()
```

**Univariate Analysis**

```python
# Distribution of numerical features
numerical_features = data.select_dtypes(include=[np.number]).columns.tolist()

for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.histplot(data[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# Distribution of categorical features (if any)
categorical_features = data.select_dtypes(exclude=[np.number]).columns.tolist()

if categorical_features:
    for feature in categorical_features:
        plt.figure(figsize=(10, 6))
        sns.countplot(x=data[feature])
        plt.title(f'Distribution of {feature}')
        plt.xticks(rotation=90)
        plt.show()
else:
    print("No categorical features found.")
```



**Bivariate Analysis**

```python

# Relationships between numerical features
plt.figure(figsize=(12, 10))
sns.heatmap(data.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot between pairs of numerical features
# Example: 'Glucose' vs 'Insulin' (replace with relevant column names)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Glucose', y='Insulin', data=data)
plt.title('Glucose vs Insulin')
plt.show()
```

**Analysis**

```python
# Example: Box plots to compare distributions across categories
# Example: 'Outcome' (replace with relevant column names)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outcome', y='Glucose', data=data)
plt.title('Glucose levels by Outcome')
plt.show()
```
