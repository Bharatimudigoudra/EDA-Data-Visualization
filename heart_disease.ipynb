*  **age**: Age of the patient.
*  **sex**: Sex of the patient (1 = male; 0 = female).
*  **cp**: Chest pain type (0: typical angina, 1: atypical angina, 2: non-anginal pain, 3: asymptomatic).
* **trestbps**: Resting blood pressure (in mm Hg).
* **chol**: Serum cholesterol in mg/dl.
* **fbs**: Fasting blood sugar > 120 mg/dl (1 = true; 0 = false).
* **restecg**: Resting electrocardiographic results (0: normal, 1: having ST-T wave abnormality, 2: showing probable or definite left ventricular hypertrophy).
* **thalach**: Maximum heart rate achieved.
* **exang**: Exercise-induced angina (1 = yes; 0 = no).
* **oldpeak**: ST depression induced by exercise relative to rest.
* **slope**: The slope of the peak exercise ST segment (0: upsloping, 1: flat, 2: downsloping).
* **ca**: Number of major vessels (0-3) colored by fluoroscopy.
* **thal**: Thalassemia (1: normal; 2: fixed defect; 3: reversible defect).
* **target**: Presence of heart disease (1 = heart disease; 0 = no heart disease).

**Violin Plots**
Visualize the distribution of numerical features across different categories using violin plots.

```python
# List of numerical features
numerical_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

# Violin plot for each numerical feature by target
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_features, 1):
    plt.subplot(2, 3, i)
    sns.violinplot(x='target', y=column, data=df, palette='Set2')
    plt.title(f'Violin Plot of {column} by Heart Disease')
plt.tight_layout()
plt.show()
```

Analysis and Insights
* Age: The distribution of age among patients shows that the presence of heart disease is relatively spread across different age groups.

* Serum Cholesterol: Higher cholesterol levels seem more common among those with heart disease, with notable outliers.
* Maximum Heart Rate Achieved: A lower maximum heart rate is more common in patients with heart disease.
* ST Depression: Higher ST depression values are more frequently observed in those with heart disease.
* Chest Pain Type: Certain types of chest pain are more associated with heart disease, such as typical angina.
