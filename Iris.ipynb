**Features of the Iris Dataset**
1.   **Id**: An identifier for each row.
2.   **SepalLengthCm**: The length of the sepal (the part of the flower that encloses the petals) in centimeters.
3.   **SepalWidthCm**: The width of the sepal in centimeters.
4.   **PetalLengthCm**: The length of the petal (the part of the flower that is often colorful and conspicuous) in centimeters.
5.   **PetalWidthCm**: The width of the petal in centimeters.
6.   **Species**: The species of the iris flower. There are three species in the dataset: Iris-setosa, Iris-versicolor, and Iris-virginica.





```python
#Load the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Iris.csv')

# Display the first few rows of the dataset
print(data.head())
```

```python
# Summary statistics of numerical features
data.describe()
```

```python
# Data types of all columns
data.dtypes
```

```python
# Checking for missing values
data.isnull().sum()
```

Visualizations to understand the distribution of the features and the relationships between them.

**Pairplot**:-
A pairplot will give us a good overview of the relationships between pairs of features and the distribution of individual features.

```python
# Pairplot
sns.pairplot(data, hue='Species', markers=["o", "s", "D"])
plt.show()
```

**Heatmap**:-
A heatmap of the correlation matrix will help us understand how the features are correlated with each other.


*   The heatmap shows the correlation coefficients between the features. A value close to 1 or -1 indicates a strong correlation, while a value close to 0 indicates a weak correlation.




```python
# Heatmap
sns.heatmap(data.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

**Boxplot**:-
Boxplots will help us see the distribution of the features and identify any potential outliers.


*   Boxplots provide a summary of the distribution of the data for each feature, including the median, quartiles, and potential outliers.


*   The boxplots reveal that Iris-setosa tends to have smaller sepal and petal measurements compared to the other species.






```python
# Boxplot to visualize the distribution of each feature
plt.figure(figsize=(12, 6))
sns.boxplot(data=data)
plt.title('Boxplot of Iris Features')
plt.show()
```

```python
# Boxplot for each feature
plt.figure(figsize=(12, 8))
sns.boxplot(data=data.drop(columns=['Id']), orient='h')
plt.title('Boxplot of Features')
plt.show()
```

**Violin Plot**:-
Violin plots will give us a better understanding of the distribution of the data for each species.


*   Violin plots combine the benefits of boxplots and kernel density plots, showing the distribution of the data across different species.



```python
# Violin plot for each feature by species
plt.figure(figsize=(12, 8))
for i, column in enumerate(data.columns[1:-1], 1):
    plt.subplot(2, 2, i)
    sns.violinplot(x='Species', y=column, data=data)
    plt.title(f'Violin Plot of {column}')
plt.tight_layout()
plt.show()
```

```python
# Scatterplot for a specific pair of features
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=data, palette='deep')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```
